```python
import pandas as pd
import numpy as np
```

```python
alarm = pd.read_csv('./data/강원도_강원도소방본부_소방용수시설.csv', encoding = 'cp949')
df = pd.read_csv('./data/구조 종별 신고.재난 데이터.csv', encoding= 'utf-8')
```

```python
df.head()
```

```python
# 데이터 화재관련 값만 추출

df = df[(df['ACDNT_CAUSE'] == '기타화재') | (df['ACDNT_CAUSE'] == '화약류') | (df['ACDNT_CAUSE'] == '방화')
        | (df['ACDNT_CAUSE'] == '폭발물') | (df['ACDNT_CAUSE'] == '화재확인')]
```

```python
df.isnull().sum()
df = df.dropna(axis = 1) # 누락값이 있는 열 삭제
```

```python
df.info()
#4097개의 행과  36개의 열
```

```python
import folium as g
from folium.plugins import MarkerCluster
```

```python
place_lat = df['LOC_INFO_X']
place_lng = df['LOC_INFO_Y']
```

```python
m = g.Map(
    location=[37.8013, 128.3745],
    zoom_start = 9
)

m.add_child(g.LatLngPopup())       # 지도에서 클릭하는 곳의 경도 위도 좌표를 표시

m
```

```python
def plot(data):
    # 리스트를 이용해 여러 행의 데이터를 위,경도로 묶음
    locations = list(zip(df.LOC_INFO_X, df.LOC_INFO_Y))

    cluster = MarkerCluster(locations = locations)
    # 지도에 클러스터를 추가.
    cluster.add_to(m)
```

```python
plot(df)
m.save('./data/화재 발생 위치.html')
```
